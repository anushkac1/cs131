```python
#file one - smallest one 

#followed code example from class
spark = SparkSession.builder.appName("MYAPP").getOrCreate()

#incorporated runtime - calculator 
start = time.time()

#read file via gsutil URL from cluster 
df = spark.read.csv("gs://dataproc-staging-us-central1-20673574438-1npjkcso/data_sample.csv", header=True, inferSchema=True)

#group by region and avergae the units sold - remember same info repeated in exact amounts (so metrics = same)
df.groupBy('Region').avg('Units Sold').show()

#end runtime 
end = time.time()

#print runtime for keeping track 
print("Runtime File 1:", round(end - start, 2), "Seconds")
```
